```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

## Graficos para energia E = 1/8

### Plot of x and y:


```python
file = pd.read_csv('E8/Data/Data_for_E8.csv')  # 1000 pontos
```

```python
for i in range(0,len(file.x),int(len(file.x)/6)):
    Y = file.y[i:i+ int(len(file.x)/6)]
    X = file.x[i:i+ int(len(file.x)/6)]
    #Energia = file.E
    plt.plot(X,Y)
    plt.title('Movimento bi-dimensional para E= 1/8 ')
    plt.text(0.4,0.5,f"$y_0$={Y[i]}",fontsize="large")
    plt.xlim(left=-0.6,right=0.6)
    plt.ylim(bottom=-0.55,top=0.65)
    plt.savefig(f"X and Y with y_0={Y[i]}.jpg",dpi=150)
    plt.show()
```

### Plot of P(y) and y:


```python
linha = [".","+",".","x",".","+"]
for i in range(6):
    file = pd.read_csv(f'E8/Data/x_zeros{i}.csv')  
    X = file.x
    Y = file.y
    P = file.p_y
    plt.plot(Y,P,f"{linha[i]}",label=f"$y_0$={Y[0]}")
    plt.xlabel('y')
    plt.ylabel('p(y)')
    plt.title('Espaco de fase para E= 1/8 ')
    plt.legend()
    plt.grid(True)
    
plt.show()
```

```python
for i in range(6):
    file = pd.read_csv(f'x_zeros{i}.csv')  
    X = file.x
    Y = file.y
    P = file.p_y
    plt.plot(Y,P,".")
    plt.xlabel('y')
    plt.ylabel('p(y)')
    plt.grid(True)
    plt.show()
```
